```python
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
text="<html><body><p>movie<p/><body/><html/>"
```

```python
remove_html_tags(text)
```

```python
df['review']=df['review'].apply(remove_html_tags)
```

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
import string
string.punctuation
```

```python
exclude=string.punctuation
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char , '')
  return text
```

```python
text= 'string. with. punctuations!#'
```

```python
remove_punc(text)
```

```python
df['review']=df['review'].apply(remove_punc)
```

```python
chat_words={'u2':'you too','bc':'bahut cute'}
```

```python
def chat_conversation(text):
  next_text=[]
  for w in text.split():
    if w.lower() in chat_words:
      next_text.append(chat_words[w.lower()])
    else:
      next_text.append(w)
  return " ".join(next_text)
```

```python
chat_conversation("tu bc hai kya")
```
