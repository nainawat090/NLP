```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'text':['people watch cricket','player was player',
                         'people write comment',
                         'cricket play comment' ] ,'output':[1,1,0,0]})
```

```python
df
```

```python
# player watch cricket player was write comment play
# [11100000]
# [00021000]
# [10000110]
# [00100011]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bow=cv.fit_transform(df['text'])
```

```python
# vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["cricket watch and write comment of cricket comment"]).toarray()
```
